```python
# Install and Import Libraries
!pip install transformers -q
!pip install datasets -q
!pip install --upgrade datasets

import pandas as pd
import torch
from datasets import load_dataset, Dataset
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, TrainingArguments, Trainer, DefaultDataCollator
from tqdm.auto import tqdm # Using tqdm.auto for better compatibility

print("Libraries installed and imported successfully.")
```

```python
# Load and Prepare Dataset
def load_and_prepare_squad(split_size=0.2, train_subset_size=5000):
    """
    Loads the SQuAD dataset, splits it into train and test sets,
    and converts the training split to a pandas DataFrame.
    """
    print(f"Loading SQuAD dataset with a train subset of {train_subset_size} samples...")
    squad = load_dataset("squad", split=f"train[:{train_subset_size}]")
    squad = squad.train_test_split(test_size=split_size, seed=42) # Added seed for reproducibility
    print("Dataset loaded and split.")
    return squad

squad = load_and_prepare_squad()
```

```python
# Initialize Tokenizer
def initialize_tokenizer(model_name="distilbert-base-uncased"):
    """
    Initializes and returns an AutoTokenizer.
    """
    print(f"Initializing tokenizer for model: {model_name}...")
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    print("Tokenizer initialized.")
    return tokenizer

tokenizer = initialize_tokenizer()
```

```python
# Preprocess Data for Training and Evaluation
def preprocess_function(examples, tokenizer, max_length=384, truncation_strategy="only_second"):
    """
    Tokenizes questions and contexts, and computes start and end positions for answers.
    This function is designed to be used with Dataset.map().
    """
    questions = [q.strip() for q in examples["question"]]
    contexts = [c.strip() for c in examples["context"]]
    inputs = tokenizer(
        questions,
        contexts,
        max_length=max_length,
        truncation=truncation_strategy,
        return_offsets_mapping=True,
        padding="max_length",
    )

    offset_mapping = inputs.pop("offset_mapping")
    answers = examples["answers"]
    start_positions = []
    end_positions = []

    for i, offset in enumerate(offset_mapping):
        answer = answers[i]
        # Handle cases where answer['text'] might be empty or contain multiple answers.
        # Assuming the first answer is the target for training.
        if not answer["text"]:
            start_positions.append(0)
            end_positions.append(0)
            continue

        start_char = answer["answer_start"][0]
        end_char = answer["answer_start"][0] + len(answer["text"][0])

        sequence_ids = inputs.sequence_ids(i)

        # Find the start and end of the context
        idx = 0
        while idx < len(sequence_ids) and sequence_ids[idx] != 1:
            idx += 1
        context_start = idx
        while idx < len(sequence_ids) and sequence_ids[idx] == 1:
            idx += 1
        context_end = idx - 1

        # If the answer is not fully inside the context, label it (0, 0)
        if (context_start >= len(offset) or context_end >= len(offset) or
            offset[context_start][0] > end_char or offset[context_end][1] < start_char):
            start_positions.append(0)
            end_positions.append(0)
        else:
            # Otherwise it's the start and end token positions
            idx = context_start
            while idx <= context_end and offset[idx][0] <= start_char:
                idx += 1
            start_positions.append(idx - 1)

            idx = context_end
            while idx >= context_start and offset[idx][1] >= end_char:
                idx -= 1
            end_positions.append(idx + 1)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs

print("Preprocessing training data...")
# Use .map for efficient preprocessing of Hugging Face Datasets
train_dataset = squad["train"].map(
    lambda examples: preprocess_function(examples, tokenizer),
    batched=True,
    remove_columns=squad["train"].column_names, # Remove original columns to keep only tokenized inputs
)
print("Training data preprocessed.")

print("Preprocessing test data...")
test_dataset = squad["test"].map(
    lambda examples: preprocess_function(examples, tokenizer),
    batched=True,
    remove_columns=squad["test"].column_names,
)
print("Test data preprocessed.")
```

```python
# Initialize Model and Data Collator
def initialize_model(model_name="distilbert-base-uncased"):
    """
    Initializes and returns an AutoModelForQuestionAnswering.
    """
    print(f"Initializing model: {model_name}...")
    model = AutoModelForQuestionAnswering.from_pretrained(model_name)
    print("Model initialized.")
    return model

model = initialize_model()
data_collator = DefaultDataCollator()
print("Data collator initialized.")
```

```python
# Configure Training Arguments and Trainer
def configure_trainer(model, tokenizer, train_dataset, eval_dataset, output_dir="qa_model"):
    """
    Configures and returns a Hugging Face Trainer.
    """
    training_args = TrainingArguments(
        output_dir=output_dir,
        eval_strategy="epoch",
        learning_rate=2e-5,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=16,
        num_train_epochs=1,
        weight_decay=0.01,
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=eval_dataset,
        tokenizer=tokenizer,
        data_collator=data_collator,
    )

    print("Trainer configured.")
    return trainer

trainer = configure_trainer(model, tokenizer, train_dataset, test_dataset)
print("Starting model training...")
trainer.train()
print("Model training complete.")
```

```python
# Evaluate a Single Instance
def evaluate_instance(instance, model, tokenizer, device):
    """
    Evaluates a single question-answering instance and returns True if the
    predicted answer matches the given answer (case-insensitive).
    """
    context = instance['context']
    question = instance['question']
    given_answer = instance['answers']['text'][0]

    inputs = tokenizer(question, context, return_tensors='pt', max_length=512, truncation=True)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        output = model(**inputs)

    start_idx = torch.argmax(output.start_logits)
    end_idx = torch.argmax(output.end_logits)

    # Convert predicted token IDs back to a string
    # Ensure end_idx is not less than start_idx
    if end_idx < start_idx:
        predicted_answer = ""
    else:
        predicted_answer_tokens = inputs['input_ids'][0][start_idx:end_idx+1] # +1 to include the end token
        predicted_answer = tokenizer.decode(predicted_answer_tokens, skip_special_tokens=True)

    return predicted_answer.lower() == given_answer.lower()
```

```python

# Calculate Accuracy on a Subset of the Dataset
def calculate_accuracy(dataset, model, tokenizer, total_count=100):
    """
    Calculates the accuracy of the model on a subset of the given dataset.
    """
    if torch.cuda.is_available():
        device = 'cuda'
        print("CUDA available, running evaluation on GPU.")
    else:
        device = 'cpu'
        print("CUDA not available, running evaluation on CPU.")

    # Move model to device
    model.to(device)
    model.eval() # Set model to evaluation mode

    correct_count = 0
    actual_total_count = min(total_count, len(dataset))
    print(f"\nEvaluating accuracy on {actual_total_count} samples...")

    for i in tqdm(range(actual_total_count)):
        try:
            if evaluate_instance(dataset[i], model, tokenizer, device):
                correct_count += 1
        except Exception as e:
            print(f"Error evaluating instance {i}: {e}")
            pass

    accuracy = correct_count / actual_total_count
    print(f'\nAccuracy: {accuracy * 100:.2f}%')
    return accuracy

# Calculate accuracy on the test set
accuracy = calculate_accuracy(squad['test'], model, tokenizer, total_count=100) # Evaluating on 100 samples from the test set
print(f'Final accuracy on the test set: {accuracy * 100:.2f}%')
```
